```python
# Extract data
import zipfile
def unzip(filename):
  zip_ref = zipfile.ZipFile(filename , "r")
  zip_ref.extractall()
  zip_ref.close()

unzip("/content/joe-biden-2020-dnc-speech.zip")
```

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/joe_biden_dnc_2020.csv')
```

```python
df.head()
```

```python
# Check the data types of all columns
data_types = df.dtypes
print(data_types)
```

```python
# Remove any unnecessary columns
train_data = df[['TEXT']].astype(str)
test_data = df[['TEXT']].astype(str)
```

```python
train_data
```

```python
# Combine the 'passage' column into a single text field
corpus = train_data['TEXT'].str.cat(sep=' ')
```

## Model

```python
# Create a Markov model
text_model = mk.Text(corpus)
```

```python
for i in range(3):
  # Generate text using the Markov model
  generated_text = text_model.make_sentence()

  # Print the generated text
  print(generated_text)
```
